## Import pandas, load data, and inspect

```python
import pandas as pd

df = pd.read_csv('datasets/worldwide_browser_trends.csv')
df.head()
```

## Check data types, convert as needed, and set index to month

```python
# check data types
print(df.info())

# convert month to datetime
df.Month = pd.to_datetime(df.Month)

# set index
df = df.set_index('Month')
df.head()
```

## Initial plot to see search interest over time by browser

```python
%matplotlib inline
df.plot()
```

## Zoom into 2011-2012: when chrome took over

```python
takeover = df.loc['2011-01':'2012-12']
takeover.plot()
```

## 6 Month rolling average plot

```python
# 6 month rolling average
rolling_six = df.rolling(window=6).mean()
rolling_six.plot(title="6 Month Rolling Avg")
```

## Quarterly growth by browser

```python
# percentage change from previous quarter
pct_change_quarterly = df.loc['2009-01':]
pct_change_quarterly = pct_change_quarterly.pct_change(periods=3) * 100
pct_change_quarterly.plot(subplots=True, figsize=(12,8))
```

## A deeper dive on chrome's performance during select years

```python
# chrome trends
chrome_trends = df.loc[df.index.year.isin([2009, 2012, 2015, 2018]), ['Google Chrome']]
chrome_trends = chrome_trends.pivot_table(
    index=chrome_trends.index.month,
    columns=chrome_trends.index.year,
    values='Google Chrome'
)
chrome_trends.plot(title="Chrome Search Performance in 2009, 2012, 2015 & 2018")
print(chrome_trends)
```
